```python
from matplotlib import pyplot as plt

# Definisco una lista di valori
slices = [60,40, 30, 40]
labels = ['sessanta', 'quaranta', 'extra1', 'extra2']

colors=['blue', 'red', 'yellow', 'green']
plt.pie(slices, labels=labels, wedgeprops={'edgecolor' : 'black'}, colors=colors)

plt.title('Pie chart di esempio')

plt.tight_layout()
plt.savefig('grafico.png')
plt.show()
```

```python
import pandas as pd
import numpy as np
from collections import Counter

df = pd.read_csv('data.csv')
ids = df['Responder_id']
lang_responses = df['LanguagesWorkedWith']

language_counter = Counter()
for response in lang_responses:
    language_counter.update(response.split(';'))

languages = []
popularity = []
for lang, ctr in language_counter.most_common(5):
    languages.append(lang)
    popularity.append(ctr)

# evidenzio il linguaggio con la percentuale più alta (quindi il primo della lista)
explode = np.zeros(len(languages))
explode[0] = 0.1                  
plt.pie(popularity, labels=languages, wedgeprops={'edgecolor' : 'black'},shadow=True, autopct='%1.1f%%', explode=explode, startangle=90)

plt.title('Popolarità linguaggi')
#plt.ylabel('Linguaggio')
plt.xlabel('Num. utilizzatori')

plt.tight_layout()
plt.savefig('grafico.png')
plt.show()
```
