```python
# Per estrarre  coloro che hanno un salario > di 70.000 $
# definisco una filter mask
high_salary = df['ConvertedComp'] > 70000

# utilizzo la mask  per filtrare i risultati desiderati (selezionando solo alcunr colonne)
df.loc[high_salary, ['Country', 'LanguageWorkedWith','ConvertedComp']]
```

```python
# Per estrarre tutte le righe che hanno nella colonna country il valore Italia o Francia
# Definisco prima la lista
countries = ['Italy', 'France']
#Poi definisco la filter mask
filt = df['Country'].isin(countries)
#Infine filtro le righe che corrispondono
df.loc[filt]
```

```python
# Per estrarre tutte le righe che hanno Python nella colonna relativa ai linguaggi di programmazione
filt = df['LanguageWorkedWith'].str.contains('Python', na=False)
df.loc[filt, 'LanguageWorkedWith']
```
