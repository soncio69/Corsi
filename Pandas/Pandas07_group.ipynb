# L'operazione di groupby prevede 3 step:
- split
- apply function
- combine result


```python
# Ottengo un DataFrameGroupBy
country_group = df.groupby(['Country'])
```

```python
# DataFrameGroupBy contiene tutti i diversi raggruppamenti x i diversi valori presenti  nella colonna indicata
country_group.get_group('Italy')
```

```python
# corrisponde a fare una ricerca con un filtro
filt = df['Country'] == 'Italy'
# df.loc[filt]
# E' successivamente possibile estrarre dal Dataframe ottenuto la colonna a cui si 
# è interessati e vedere le occorrenze presenti
df.loc[filt]['SocialMedia'].value_counts()
```

```python
# Grazie a groupby è possibile applicare la funzione a tutti i raggruppamenti presenti nel DataFrame (e non solo a uno)
# Si deve utilizzare il DataFrameGroupBy ottenuto col groupby
#In questo modo si ottiene una Series con 2 indici (Country e SocialMedia)
country_group['SocialMedia'].value_counts().head(50)
```

```python
# E' possibile accedere ad uno dei raggruppamenti della serie accedendo con l'indice
country_group['SocialMedia'].value_counts().loc['India']
```

```python
# Per avere il salario medio nei diversi  paesi.
# Raggruppo per la colonna 'Country' e applico la funzione median sul DataFrameGroupBy ottenuto
country_group = df.groupby(['Country'])
country_group['ConvertedComp'].median()
```

```python
# Viene restituito  una Series indicizzata x Country 
country_group['ConvertedComp'].median().loc['Italy']
```

```python
# E' possibile passare più aggregate function contemporaneamente utilizzando  il metodo agg() e passando le
# diverse funzioni da eseguire sui diversi raggruppamenti
country_group['ConvertedComp'].agg(['mean', 'median'])
```

```python
country_group['ConvertedComp'].agg(['mean', 'median']).loc['France']
```

```python
# Per estrarre il numero di persone che conoscono Python nei diversi Country
# Vediamo prima come estrarre per un solo paese utilizzando un  filtro 
filt = df['Country'] == 'Italy'
df.loc[filt]['LanguageWorkedWith'].str.contains('Python')
```

```python
# Utilizzando il metodo sum() con dei boolean, viene sommato 1 x True e 0 per False.
# Quindi per avere il numero di chi usa Python si può faredf.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
df.loc[filt]['LanguageWorkedWith'].str.contains('Python').sum()
```

```python
# NOn è possibile fare (come prima):
# country_group['LanguageWorkedWith'].str.contains('Python').sum()
# in quanto SeriesGroupBy' object has no attribute 'str' (come invece ha Series)
# E' necessario utilizzare il metodo apply()
```

```python
# Si deve passare al metodo apply una lambda che applica il medoto a tutte le Series che fanno parte della SeriesGroupBy
country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
# Per ottenere la percentuale di chi  utilizza python  sul numero  di risposte per ciascun  paese
# ottengo prima il numero di risposte x paese
df_respondent = df['Country'].value_counts()
df_respondent
```

```python
# Poi ottengo il numero di utilizzatori di python (vedi punto precedente) e li memorizzo in un DataFrame
df_use_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
df_use_python
```

```python
# Per concatenare 2 Series (o DataFrame) in un unico DataFrame si utilizza il metodo concat
# Per unire le 2 colonne devo passare axis='columns'. 
# Diversamente verrebbero  unite le rows (default)
python_df = pd.concat([df_respondent,df_use_python], axis='columns', sort=False)
python_df
```

```python
# E' preferibile rinominare le colonne
python_df.rename(columns={'Country' : 'Totale', 'LanguageWorkedWith' : 'Python'}, inplace=True)
python_df
```

```python
python_df# Creo la nuova colonna con la percentuale degli  utilizzatori di python sul totale
python_df['PctKnowPython'] = (python_df['Python']/python_df['Totale'])*100
python_df
```

```python
# Per ordinarli
python_df.sort_values(by='PctKnowPython', ascending=False, inplace=True)
```
